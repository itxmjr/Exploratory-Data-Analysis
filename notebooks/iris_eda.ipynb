# ðŸ“Š Iris Dataset - Exploratory Data Analysis

## Introduction

The **Iris dataset** is one of the most famous datasets in machine learning and statistics. It was introduced by British statistician and biologist Ronald Fisher in 1936.

### Dataset Overview:
- **150 samples** of iris flowers
- **3 species**: Setosa, Versicolor, and Virginica (50 samples each)
- **4 features**: Sepal length, Sepal width, Petal length, and Petal width (all in cm)

## 1. Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set style for better-looking plots
sns.set_style("whitegrid")
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (10, 6)

# Display settings
pd.set_option('display.max_columns', None)

print("Libraries imported successfully!")
```

## 2. Load Dataset

```python
iris_df = sns.load_dataset("iris")
print("Dataset loaded successfully!")
print(f"\nDataset shape: {iris_df.shape}")
```

## 3. Data Inspection

Let's examine the structure and quality of our dataset.

```python
print("First 5 rows of the dataset:\n")
iris_df.head()
```

```python
print("Dataset Information:\n")
iris_df.info()
```

```python
print("Missing Values Check:\n")
print(iris_df.isnull().sum())
print(f"\nTotal missing values: {iris_df.isnull().sum().sum()}")
```

## 4. Statistical Summary

```python
print("Overall Statistical Summary:\n")
iris_df.describe()
```

```python
# Statistics grouped by species
print("Statistical Summary by Species:\n")
iris_df.groupby('species').describe()
```

## 5. Data Visualizations

### 5.1 Scatter Plots - Feature Relationships

```python
# Petal Length vs Petal Width
plt.figure(figsize=(5, 3))
for species in iris_df['species'].unique():
    data = iris_df[iris_df['species'] == species]
    plt.scatter(data['petal_length'], data['petal_width'], 
               label=species, alpha=0.7, s=50, edgecolors='black', linewidth=0.5)

plt.xlabel('Petal Length (cm)', fontsize=10)
plt.ylabel('Petal Width (cm)', fontsize=10)
plt.title('Petal Length vs Petal Width by Species', fontsize=12, fontweight='bold', pad=15)
plt.legend(title='Species', fontsize=8)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nðŸ“Œ Key Insight: Strong linear relationship between petal dimensions, especially for Versicolor and Virginica")
```

```python
# Sepal Length vs Sepal Width
plt.figure(figsize=(5, 3))
for species in iris_df['species'].unique():
    data = iris_df[iris_df['species'] == species]
    plt.scatter(data['sepal_length'], data['sepal_width'], 
               label=species, alpha=0.7, s=50, edgecolors='black', linewidth=0.5)

plt.xlabel('Sepal Length (cm)', fontsize=10)
plt.ylabel('Sepal Width (cm)', fontsize=10)
plt.title('Sepal Length vs Sepal Width by Species', fontsize=12, fontweight='bold', pad=15)
plt.legend(title='Species', fontsize=8)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nðŸ“Œ Key Insight: Setosa shows different sepal characteristics with higher width relative to length")
```

### 5.2 Pair Plot - Comprehensive Feature Relationships

```python
plt.figure()
pair_plot = sns.pairplot(iris_df, hue='species', markers=['o', 's', 'D'],
                         diag_kind='kde', height=2.5, aspect=1.2)
pair_plot.fig.suptitle('Pairwise Relationships of Iris Features by Species', 
                        y=1.02, fontsize=16, fontweight='bold')
plt.show()

print("\nðŸ“Œ Key Insights:")
print("- Setosa (blue) is clearly separated from other species")
print("- Versicolor (orange) and Virginica (green) show some overlap")
print("- Petal measurements provide better species separation than sepal measurements")
```

### 5.3 Box Plots

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

fig, axes = plt.subplots(2, 2, figsize=(6, 6))
axes = axes.ravel()

for idx, feature in enumerate(features):
    sns.boxplot(data=iris_df, x='species', y=feature, ax=axes[idx], palette='pastel', hue="species", legend=False)
    axes[idx].set_title(f'{feature.replace("_", " ").title()} by Species', 
                       fontsize=12, fontweight='bold')
    axes[idx].set_xlabel('Species')
    axes[idx].set_ylabel(f'{feature.replace("_", " ").title()} (cm)')

plt.suptitle('Box Plots - Feature Comparison by Species', fontsize=16, fontweight='bold', y=1.00)
plt.tight_layout()
plt.show()

print("\nðŸ“Œ Key Insight: Few outliers present, indicating good data quality")
```
